SEGUNDO DESAFIO

Precio del metro cuadrado por barrio, graficar el resultado segun la informacion sea visble

*Calculamos el precio del m2 por barrio en la ciudad de Bogotá genera una nueva *

```python
precio_m2 = inmuebles['Precio_Millon'] / inmuebles['Area']
precio_m2_barrio = precio_m2.groupby(inmuebles['Barrio']).mean().sort_values(ascending=False).head(10)
print('Precio de m2 por', precio_m2_barrio)
```

```python

#Importamos las librerías
import matplotlib.pyplot as plt
import seaborn as sns

#Clonfiguramos el tamaño del lienzo
plt.subplots(figsize=(8,4))

#Definimos la paleta de colores para cada tipo de inmueble
color_barra = plt.get_cmap('tab10').colors[:len(precio_m2_barrio)]

#Creamos el gráfico de barras
plt.bar(precio_m2_barrio.index, precio_m2_barrio, color=color_barra)

#Estilamos los títulos y las etiquetas
plt.title('Precio del m2 por barrio en la ciudad de Bogotá', fontsize=16)
plt.xlabel('Barrio', fontsize=11)
plt.ylabel('Precio del m2 por barrio en $', fontsize=12)
plt.xticks(rotation=50, fontsize=10)
#Código para configuración para guardado de la gráfica en algún formato
plt.savefig('/content/drive/MyDrive/inversion_datos/valor_barrios_m2.png', format='png')
plt.show()
```

```python
# Calculamos el Precio por metro cuadrado y lo agregamos a Precio/m2
inmuebles['Precio/m2_millon'] = inmuebles['Precio_Millon']/inmuebles['Area']
inmuebles[['Tipo','Precio','Area','Precio/m2_millon']].sample(7)
```

```python
# Calcular el precio promedio por cada m2 en cada barrio.
precio_m2_media_barrio = inmuebles.groupby('Barrio')['Precio/m2_millon'].mean().reset_index()
sns.barplot(x='Precio/m2_millon', y='Barrio', data=precio_m2_media_barrio.head(20))

# Etiqueta al grafico
plt.title('Precio por metro cuadrado para venta de inmuebles')
plt.xlabel('Precio por metro cuadrado ("En millones")')
plt.ylabel('Barrio')
plt.show()
```

```python
inmuebles.groupby('Barrio').mean()
```

```python
datos_barrio = inmuebles.groupby('Barrio').sum()
datos_barrio
```

```python
datos_barrio['valor_m2_Barrio'] =datos_barrio['Precio_Millon']/datos_barrio['Area']
datos_barrio
```

```python
m2_barrio = dict(datos_barrio['valor_m2_Barrio'])
```

```python
inmuebles['Valor_m2_Barrio'] = inmuebles['Barrio']
inmuebles['Valor_m2_Barrio'] = inmuebles['Valor_m2_Barrio'].map(m2_barrio)
inmuebles.head(5)
```

```python
top_barrios = inmuebles['Barrio'].value_counts()[:10].index
```

```python
datos_barrio.reset_index(inplace=True)
datos_barrio
```

```python
datos_barrio.query('Barrio in @top_barrios')
```

Además del gráfico de barras podemos tener la descripción estadística, para eso usamos bloxplot (gráfico de cajas) y eliminamos sns.barplot

```python
plt.figure(figsize=(10,8))
#Además del gráfico de barras podemos tener la descripción estadística, para eso usamos bloxplot (gráfico de cajas) y eliminamos sns.barplot
ax = sns.boxplot(x='Barrio', y='valor_m2_Barrio',data = datos_barrio.query('Barrio in @top_barrios'))
ax.tick_params(axis='x', rotation=45)
plt.show()     
```

```python
#De aciuerdo con el gráfico anterior, no tiene mucho sentido verificar el comportamiento del área, por lo mismo mejor lo cambiamos a precios.
plt.figure(figsize=(10,8))
#Además del gráfico de barras podemos tener la descripción estadística, para eso usamos bloxplot (gráfico de cajas) y eliminamos sns.barplot
#Se calcula sobre inmuebles totales y no datos barrio, ya que los datos sin insuficientes.
ax = sns.boxplot(x='Barrio', y='valor_m2_millon',data = inmuebles.query('Barrio in @top_barrios'))
ax.tick_params(axis='x', rotation=45)
plt.show()
#Los outlayes son los datos discrepantes, muy fuera del comportamiento normal
```
